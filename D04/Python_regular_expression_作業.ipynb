### 讀入資料文本

```python
#讀取文本資料
f=open('sample_emails.txt', mode='r')
sample_corpus=f.read()
```

### 讀取寄件者資訊
觀察文本資料可以發現, 寄件者資訊都符合以下格式

`From: <收件者姓名> <收件者電子郵件>`

```python
import re
ru=r'From: "[^"]*" <[^>]*>'
match=re.findall(ru,sample_corpus)
```

```python
match
```

### 只讀取寄件者姓名

```python
ru=r'(?<=From: )"[^"]*"'
for name in re.findall(ru,sample_corpus):
    print(name)
```

### 只讀取寄件者電子信箱

```python
ru=r'(?<=" <)\w*@[^>]*'
mails=re.findall(ru,sample_corpus)
for mail in mails:
    print(mail)
```

### 只讀取電子信箱中的寄件機構資訊
ex: james_ngola2002@maktoob.com --> 取maktoob

```python
ru=r'(?<=" <)\w*@[^>]*'
for mail in re.findall(ru,sample_corpus):
    print(re.search('(?<=@)\w*',mail).group())
```

### 結合上面的配對方式, 將寄件者的帳號與機構訊返回
ex: james_ngola2002@maktoob.com --> [james_ngola2002, maktoob]

```python
for mail in mails:
    acc = re.search('\w*(?=@)*',mail).group()
    mx =  re.search('(?<=@)\w*',mail).group()
    print('{}, {}'.format(acc , mx))
```
